```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as MAE
import xgboost
import tensorflow as tf
```

```python
shuffled_df=df.sample(frac=1,ignore_index=True)
```

```python
dataset=np.array(shuffled_df)
dataset.shape
```

```python
X = dataset[:,:9]
y = dataset[:,9:]
y_co = y[:,0]
y_nox = y[:,1]
X.shape , y.shape , y_co.shape , y_nox.shape
```

```python
#generating Train, Test and Valid dataset and checking their shapes

X_train_full , X_test , y_train_full , y_test = train_test_split(X , y, train_size=0.85)
X_train , X_valid , y_train , y_valid = train_test_split(X_train_full , y_train_full)

X_train.shape, X_valid.shape, X_test.shape
```

```python
# creating "xgboost model" with defualt hyperparameters
# creating scaler using StandardScaler to normalize data

model_co = xgboost.XGBRegressor()
model_nox = xgboost.XGBRegressor()
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled = scaler.transform(X_test)
```

```python
# Separeting labels to use each one in individual training model
y_co_train = y_train[:,0]
y_co_valid = y_valid[:,0]
y_co_test = y_test[:,0]

y_nox_train = y_train[:,1]
y_nox_valid = y_valid[:,1]
y_nox_test = y_test[:,1]
```

```python
y_co_train.shape
```

```python
model_co.fit(X_train_scaled, y_co_train, eval_set=[(X_valid_scaled, y_co_valid)], early_stopping_rounds=5)
```

```python
model_nox.fit(X_train_scaled, y_nox_train, eval_set=[(X_valid_scaled, y_nox_valid)], early_stopping_rounds=5)
```

```python
y_co_pred = model_co.predict(X_test_scaled)
y_nox_pred = model_nox.predict(X_test_scaled)
```

```python
# Evaluting models

print('Mean Absolute Error for CO prediction: ',MAE(y_co_test, y_co_pred))
print('Mean Absolute Error for Nox prediction: ',MAE(y_nox_test, y_nox_pred))
```
